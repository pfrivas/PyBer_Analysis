```python
import matplotlib.pyplot as plt
```

```python
# Setting the axis

savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Assign the Variables
plt.plot(months, savings)
    # Order of arguments will matter
        # X variables in the first position
        # Y variables in the second position 

# Label the Axis
plt.xlabel('Month')
plt.ylabel('Savings in Dollars')
plt.title('Saving Account Results')

# Show the Plot
plt.show()
    # this should always be the last line of code
```
